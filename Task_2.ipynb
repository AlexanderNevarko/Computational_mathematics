# Практическая задача №2 от Александра Неварко
Решение д/у методом стрельбы.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


# Block of conastants
N = 1000 # Increase of this number after 1000 does not improve the accuaracy
x0 = 0
xN = 1
y0 = 1
yN = np.e

# Preparing x, y, p
x = np.linspace(x0, xN, N)
y, p = np.zeros(N), np.zeros(N)
h = (xN - x0) / (N-1)
y[-1] = yN


# Get the value of y[0] having p[N-1] = a
# Also changes the y[] array
def simulate(a, x, y, p, N):
    p[-1] = a
    for n in range(N-2, -1, -1):
        y[n] = y[n+1] - h*p[n+1]
        p[n] = h*y[n+1] + (1 + h * np.exp(x[n]))*p[n+1] - h*(2*np.exp(x[n]) + np.exp(2*x[n]))
    return y[0]

simulate(2.7, x, y, p, N)
plt.figure(dpi = 80, figsize = (10, 5))
plt.plot(x, y, color = 'blue')
simulate(2.75, x, y, p, N)
plt.plot(x, y, color = 'red')
```

Now, when the value of right derivative is localized between <br>
`2.70 and 2.75` <br>
we can search the right value, that will fit the task


```python
M = 1000
a = np.linspace(2.7, 2.75, M)
sim_y = np.zeros(1000)
for i in range(M):
    sim_y[i] = simulate(a[i], x, y, p, N)
    
plt.figure(dpi = 80, figsize = (10, 5))
plt.plot(a, sim_y, color = 'blue')
```

Now we can see, that the maximum accuracy, that we can achieve is approximately<br>
`(1.04 - 0.92) / 1000 / 1 = 1.2e-5`<br>
Now let's find the right alpha

```python
eps = 7e-5
alpha = []
for i in range(M):
    if abs(sim_y[i] - 1.0) <= eps:
        alpha.append(a[i])
print(alpha)
```

```python
alpha = alpha[0]
```

## The resulting solution

```python
simulate(alpha, x, y, p, N)
plt.figure(dpi = 80, figsize = (10, 5))
plt.plot(x, y, color = 'red')
```
